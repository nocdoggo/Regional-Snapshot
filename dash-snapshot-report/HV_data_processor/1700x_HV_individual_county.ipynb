```python
df_19 = pd.read_csv('2019County.csv')
df_19['Fiscal Year'] = 2019
df_19
```

```python
df_20 = pd.read_csv('2020County.csv')
df_20['Fiscal Year'] = 2020
df_20
```

```python
nt = pd.concat([df_19, df_20], axis = 0)
con = nt.reset_index().drop(columns = 'index')
con1 = con[(con['County'] != "Illinois") & (con['County'] != "Cook")].reset_index().drop(columns = 'index')
con1
```

```python
con1['County'].unique()
```

```python
code = []
for ser in con1['GEOID']:
    code.append(ser[-3:])
code
```

```python
con1['num'] = code
con1
subset = con1.drop(columns = 'GEOID')
subset
```

```python
county = subset['County'].unique()
code1 = subset['num'].unique()
for i, r in zip(code1, county):
    col = subset.columns[1:9]
    my_dic = {}
    n = 0
    for c in col:
        my_dic[n] = c
        n = n+1
    my_dic[0] = "/"
    my_list =['Fiscal Year', 'Proposed Capacity', 'Funded Enrollment', 'Funded Enrollment', 'Enrollment',
             'Enrollment', 'Enrollment', 'Enrollment']
    my_dic0 = {}
    n_0 = 0
    for l in my_list:
        my_dic0[n_0] = l
        n_0 += 1
    a1 = subset[(subset['County'] == r)& (subset['Fiscal Year'] == 2019)]
    row_19 = a1.iloc[0, 1:9].tolist()
    my_dic1 = {}
    n_1 = 0
    for v1 in row_19:
        my_dic1[n_1] = v1
        n_1 += 1
    a2 = subset[(subset['County'] == r)& (subset['Fiscal Year'] == 2020)]
    row_20 = a2.iloc[0, 1:9].tolist()
    my_dic2 = {}
    n_2 = 0
    for v2 in row_20:
        my_dic2[n_2] = v2
        n_2 += 1
    # csv header
    fieldnames = [0, 1, 2, 3, 4, 5, 6, 7]
    #csv header
    rows = [my_dic, my_dic0, my_dic1, my_dic2]
    with open((str(i) + str('_HV')+'.csv'), 'w', encoding='UTF8', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)
```
